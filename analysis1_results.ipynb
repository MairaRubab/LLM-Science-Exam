```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data_target = pd.read_csv("C:/Users/maira/OneDrive/Desktop/Kaggle Project/train.csv")
data_predictions = pd.read_csv("C:/Users/maira/OneDrive/Desktop/Kaggle Project/output1_temp1.csv")
data_target.shape
data_predictions.shape
len(data_target) == len(data_predictions)
#assert len(data_target) == len(data_predictions)
diff_count = (data_target['answer'] != data_predictions['output1']).sum()
print(f"Number of differing entries: {diff_count}")  ### 62% answers are correct
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
# Extract relevant columns
true_answers = data_target['answer']
predicted_answers = data_predictions['output1']
# Calculate accuracy
accuracy = accuracy_score(true_answers, predicted_answers)
print(f"Accuracy: {accuracy:.2f}")
# Calculate confusion matrix
conf_matrix = confusion_matrix(true_answers, predicted_answers, labels=['A', 'B', 'C', 'D', 'E'])
conf_matrix_display = ConfusionMatrixDisplay(conf_matrix, display_labels=['A', 'B', 'C', 'D', 'E'])
# Plot confusion matrix
conf_matrix_display.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report
# Generate classification report
report = classification_report(true_answers, predicted_answers, labels=['A', 'B', 'C', 'D', 'E'])
print("Classification Report:")
print(report)
```
